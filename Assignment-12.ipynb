# Removing the duplicates from a list while preserving the original order ?

```python
def remove_duplicates(lst):
    seen = set()
    return [x for x in lst if not (x in seen or seen.add(x))]
print(remove_duplicates([1, 2, 2, 3, 1, 4]))
```

# Merging the two dictionaries into one

```python
def merge_dicts(d1, d2):
    return {**d1, **d2}
print(merge_dicts({'a':1}, {'b':2}))
```

# Determine if a number is prime number or not

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
```

# Return all even numbers from a list

```python
def find_evens(lst):
    return [x for x in lst if x % 2 == 0]
print(find_evens([1, 2, 3, 4, 5, 6]))
```

# Finding the missing number from 1 to n

```python
def find_missing(nums, n):
    return n * (n+1)//2 - sum(nums)
print(find_missing([1, 2, 4, 5], 5))
```

# Check if two lists are equal or not

```python
def lists_equal(l1, l2):
    return l1 == l2
print(lists_equal([1, 2, 3], [1, 2, 3])) 
```

# Return common elements without duplicates or find intersection of two lists

```python
def intersection(l1, l2):
    return list(set(l1) & set(l2))
print(intersection([1, 2, 3, 4], [3, 4, 5]))
```

# Count how many times an element appears in a list

```python
def count_occurences(lst, x):
    return lst.count(x)
print(count_occurences([1, 2, 2, 3, 2], 2))
```

# Find the second largest element in a list

```python
def second_largest(lst):
    return sorted(set(lst))[-2]
print(second_largest([10, 20, 4, 45, 99]))
```

# Finding GCD of two numbers

```python
import math
def gcd(a, b):
    return math.gcd(a, b)
print(gcd(20, 28))
```
