# Find longest word in a sentence

```python
def longest_word(s):
    words = s.split()
    return max(words, key = len)
print(longest_word("Python coding interviews are easy"))
```

# Determine whether a number is a power of two or not

```python
def is_power_of_two(n):
    return n > 0 and (n & (n-1)) == 0
print(is_power_of_two(16))
print(is_power_of_two(18))
```

# Remove all whitespaces from string

```python
def remove_spaces(s):
    return "".join(s.split())
print(remove_spaces("Py th on \n"))
```

# Find First Non-Repeating character

```python
from collections import Counter
def first_unique(s):
    freq = Counter(s)
    for c in s:
        if freq[c] == 1:
            return c
    return None
print(first_unique("swiss"))
```

# Rotate list to the right by k steps

```python
def rotate_list(lst, k):
    k %= len(lst)
    return lst[-k:] + lst[:-k]
print(rotate_list([1, 2, 3, 4, 5], 2))
```

# Find string length without using len()

```python
def str_length(s):
    count = 0
    for _ in s:
        count += 1
    return count
print(str_length("python"))
    
```

# Move all zeros to the end while keeping order

```python
def move_zeros(lst):
    return [x for x in lst if x != 0] + [0] * lst.count(0)
print(move_zeros([0, 1, 0, 3, 12]))
```

# Retun min and max from a list

```python
def min_max(lst):
    return min(lst), max(lst)
print(min_max([4, 1, 9, 2]))
```
