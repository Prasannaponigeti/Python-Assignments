## 1. Finding the Maximum Number in a list

```python
list = [5, 23, 44, 41, 26, 6, 9, 11, 55]
max = 0
for num in list:
    if max < num:
        max = num
print(max)
```

## 2. Finding the Minimum Number in a list

```python
list = [5, 23, 44, 41, 26, 6, 9, 11, 55]
min = 0
for num in list:
    if min > num:
        min = num
print(min)
```

## 3. Finding the Middle Element in a list

```python
list = [5, 23, 44, 41, 26, 6, 9, 11, 55] 
midelement = int((len(list)/2))
print(list[midelement])
```

## 4. converting a list into a string

```python
list = ["P", "Y", "T", "H", "O", "N"]
string = "".join(list)
print(string)
print(type(string))
```

## 5. Comparing Two strings for ANAGRAMS

```python
str1 = "listen"
str2 = "silent"
if sorted(str1) == sorted(str2):
    print("True")
else:
    print("False")
```

## 6. Checking for PALINDROME using extended slicing technique

```python
str1 = "kayak"
if str1 == str1[::-1]:
    print("True")
else:
    print("False")
    
```
