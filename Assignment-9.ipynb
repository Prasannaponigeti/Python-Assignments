# Implementing variable length arguments in python

```python
def average(*t):
    avg = sum(t)/len(t)
    return avg
result1 = average(32, 4, 56, 78, 23, 34, 57, 64)
result2 = average(5, 10, 15, 20, 25, 30, 35, 40, 45, 50)
print("Average is:" , result1)
print("Average is :", result2)
```

# Creating instance member variable in python


```python
class Test:
    def __init__(self):
        self.a = 5
    def f1(self):
        self.b = 10
t1 = Test()
t2 = Test()
t1.f1()
t1.c = 15
print(t1.__dict__)
print(t2.__dict__)
```

# Addition using Lambda functions

```python
f = lambda a, b : a+b
r = f(3,7)
print(r)
```

# Finding factorial using lambda function

```python
f = lambda n:1 if n == 0 else n * f(n-1)
f(5)
```

# simple python patterns

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(j, end = "")
    print()
```

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(i, end = "")
    print()
```

```python
for i in range(1, 6):
    for j in range(1, 6):
        print("*", end = "")
    print()
```

```python
for i in range(1, 6):
    for j in range(1, 6):
        print(j%2, end = "")
    print()
```
